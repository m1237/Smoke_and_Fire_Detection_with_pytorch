```python
# Extract All Video Frames

import cv2
import os


print(cv2.__version__)
vidcap = cv2.VideoCapture('C:/Users/PC/Desktop/Dataset_FNS/Camlıca-Duman/Duman-18.09.2019/Duman1-18.09.2019/Duman1-18.09.2019/1# CT-CC-KMS-57-CIKIS@2019-09-18T201058.992Z.mkv')
success,image = vidcap.read()
count = 0
success = True

os.chdir('C:/Users/PC/Desktop/Dataset_FNS/Dataset/Smoke')
while success:
    cv2.imwrite("Frame_0_%d.jpg" % count, image)     # save frame as JPEG file
    success,image = vidcap.read()
    print("Read a new frame: ", success)
    count += 1
```
